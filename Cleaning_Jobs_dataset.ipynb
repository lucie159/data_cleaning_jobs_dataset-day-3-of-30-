#### COLONE Salary Estimate

```python

#function which extract numbers and hyphen in Salary Estimate column
def extract_number(salary):
    return re.sub(r'[^\d\-]',"", str(salary))
df['Salary Estimate'] = df['Salary Estimate'].apply(extract_number)
df
```

```python
def extract_salary_range(salary_str):
    if "-" in salary_str:
        min_salary, max_salary = map(int, salary_str.split("-"))
    else:
        min_salary = max_salary = int(salary_str)
    return min_salary, max_salary


   
df[["min_salary", "max_salary"]] = df["Salary Estimate"].apply(extract_salary_range).tolist()
df

```

```python
# Let find the average salary
df['average'] = df.apply(lambda x: (x['min_salary'] + x['max_salary'])/2, axis=1)
df.head()
```

#### COLUMN COMPANY NAME



```python
#We extract the text that comes before the \n 
def extract_before_newline(text):
    
    if '\n' in text:
        return text.split('\n')[0]
    else:
        return text
df["Company Name"] = df["Company Name"].apply(extract_before_newline)
df
```

#### COLUMN Rating 

```python
df["Rating"].unique()
```

```python
# We normalise the values, that is all values less than 0 are brought to 0

def normalizeRating(value):
     if value is None:
        return None
     elif value < 0:
        return 0
     else:
        return value

df["Rating"] = df["Rating"].apply(normalizeRating)
df
```

#### Column Company age

```python
#To know the company name, we subtract the foundation date from 2020.
DATE = 2020
def companyAge(date):
    if date == -1:
        return date
    else:
        age = DATE - date
        return age

df["Company age"] = df["Founded"].apply(companyAge)
df
    
    
```

```python
#Extracting the jobs in the Job title
def job_extraction(job):
  if 'data scientist' in job.lower():
    return 'data_scientist'
  elif 'analyst' in job.lower():
    return 'analyst'
  elif 'machine learning engineer' in job.lower():
    return 'mle'
  else:
    return 'na'

df['job_simp'] = df['Job Title'].apply(job_extraction)
df
```

```python
# Create a dictionary of specialties and their corresponding column names
specialties = {
    'python': 'python',
    'excel': 'excel',
    'hadoop': 'hadoop',
    'spark': 'spark',
    'aws': 'aws',
    'tableau': 'tableau',
    'big_data': 'big_data'
}

# Loop through the dictionary and create the columns
for specialty, column_name in specialties.items():
    df[column_name] = df['Job Description'].apply(lambda x: 1 if specialty.lower() in str(x).lower() else 0)

df
```

```python
# Drop rows with -1 in any of the specified columns

df = df[~df[['Headquarters', 'Size', 'Founded', 'Revenue','Type of ownership','Industry', 'Sector', 'Job Title','Salary Estimate', 'Rating']].isin([-1]).any(axis=1)]
df
```

```python
#we verified if the value -1 exist in dataset
df.isin([-1]).sum().sum()
```

#### COLUMN Size

We will extract the Minimum and the maximun number of employees

```python
df['Size'].unique()
```

```python
def extract_min_max(size_str):
    if 'to' in size_str:
        min_val, max_val = size_str.split(' to ')
        min_val = int(min_val)
        max_val = int(max_val.split(' ')[0])
    elif '+' in size_str:
        min_val = int(size_str.split('+')[0])
        max_val = float('inf')
    else:
        min_val = max_val = None
    
    return pd.Series([min_val, max_val])

# Apply the extract_min_max function to the 'Size' column
df[['Min_Employees', 'Max_Employees']] = df['Size'].apply(extract_min_max)
df
```

```python
#We check if any nan value has been created
df["Min_Employees"].isna().sum()
```

#We drop this nan values
df = df.dropna(subset=['Min_Employees'])
df["Min_Employees"].isna().sum()

```python
df["Max_Employees"].isna().sum()
```

#### Column Revenue

```python
df["Revenue"].isna().sum()
```

```python
# Remove the dollar sign and extract the numeric part of the 'Revenue' column
df['Revenue'] = df['Revenue'].str.replace('$', "").str.extract(r'(\d+(?:\.\d+)?)', expand=False).astype(float)

# Define a function to handle the ranges
def handle_range(revenue):
    if isinstance(revenue, str) and 'to' in revenue:
        values = [float(x) for x in revenue.split('to')]
        return (values[0] + values[1]) / 2
    else:
        return revenue

# Apply the handle_range function to the 'Revenue' column
df['Revenue'] = df['Revenue'].apply(handle_range)
df
```

```python
#Check for nan values
df["Revenue"].isna().sum()
```

```python
#We replace the nan values by the mean revenue

mean_revenue = df["Revenue"].mean()
df['Revenue'].fillna(mean_revenue, inplace=True)
df["Revenue"].isna().sum()
```

```python
#We drop the columns index and Competitors
column_list = ['index', 'Competitors']
df.drop(column_list, axis=1)
```
